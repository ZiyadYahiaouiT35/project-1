```python
import matplotlib.pyplot as plt
import numpy as np

models = ['Random Forest', 'XGBoost', 'LightGBM', 'Logistic Regression']
metrics = ['Accuracy', 'F1 Score', 'ROC AUC']
scores = {
    'Accuracy': [0.902, 0.890, 0.878, 0.793],
    'F1 Score': [0.902, 0.890, 0.878, 0.792],
    'ROC AUC': [0.967, 0.958, 0.961, 0.912]
}
x = np.arange(len(metrics))  
width = 0.2 
fig, ax = plt.subplots(figsize=(12, 7))
colors = ['#6d8dc0', '#e0966d', '#78b98a', '#c76b73'] 
for i, model in enumerate(models):
    model_scores = [scores[metric][i] for metric in metrics]
    rects = ax.bar(x + (i - 1.5) * width, model_scores, width, label=model, color=colors[i])
    ax.bar_label(rects, padding=3, fmt='%.3f', fontsize=10)
ax.set_ylabel('Score', fontsize=12)
ax.set_title('Comparaison des modèles — Heart Failure Prediction', fontsize=14, pad=20)
ax.set_xticks(x)
ax.set_xticklabels(metrics, fontsize=12)
ax.legend(loc='lower right')
ax.yaxis.grid(True, linestyle='--', alpha=0.7)
ax.set_ylim(0, 1.1)
plt.tight_layout()
plt.savefig('../comparaison_modeles_final.png', dpi=300)
plt.show()
```
